```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

### Вичитування зображення та переведення в RGB

```python
img = cv.imread("image.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

### Поділ на канали

```python
red, green, blue = cv.split(img)
```

### Створення зображень в RGB, RBG, GRB, GBR

```python
rgb = cv.merge([red, green, blue])
rbg = cv.merge([red, blue, green])
grb = cv.merge([green, red, blue])
gbr = cv.merge([green, blue, red])
```

### Обʼєднання оотриманих зображень в колаж

```python
half1 = np.hstack([rgb, rbg])
half2 = np.hstack([grb, gbr])
collage = np.vstack([half1, half2])
```

```python
plt.imshow(collage)
plt.show()
```
